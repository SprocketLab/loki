```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot

from tqdm import tqdm
```

```python
with open("lshtc/train-remapped.csv", "r") as f:
    lines = f.readlines()

class_set = set()
labels = []
features = []
for l, line in tqdm(enumerate(lines), total=len(lines) - 1):
    # if l > 100: break
    if l == 0: continue
    line = line.strip().split(" ")
    label = []
    feature = {}
    for element in line:
        if ":" not in element:
            element = int(element.replace(",", ""))
            class_set.add(element)
            label.append(element)
        else:
            feature_id = int(element.split(":")[0])
            feature_value = int(element.split(":")[1])
            feature[feature_id] = feature_value
    labels.append(label)
    features.append(feature)
```

```python
labelspace = set([val for sublist in labels for val in sublist])
len(labelspace)
labelspace
```

```python
with open("lshtc/hierarchy.txt", "r") as f:
    lines = f.readlines()

G = nx.Graph()

for l, line in tqdm(enumerate(lines), total=len(lines) - 1):
    a, b = line.split(' ')
    a = int(a.strip())
    b = int(b.strip())
    if a in labelspace or b in labelspace:
        if a not in G.nodes():
            G.add_node(a)
        if b not in G.nodes():
            G.add_node(b)
        G.add_edge(a, b)
```
